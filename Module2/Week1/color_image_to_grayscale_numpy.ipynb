# **Description**

Ảnh sẽ được biểu diễn bởi tập các điểm ảnh (Pixel) có giá trị từ 0 đến 255. Ảnh màu sẽ được biểu diễn thành ma trận có kích thước (Height, Width, Channel); trong đó số Channel là 3 tương ứng là giá trị biểu diễn cho các kênh màu R-Red, G-Green, B-Blue. Ảnh xám sẽ được biểu diễn thành ma trận
có kích thước (Height, Width). Vì vậy, để chuyển ảnh màu thành ảnh xám, chúng ta sẽ tổng hợp từ 3 kênh màu RGB thành 1 giá trị duy nhất trong khoảng 0 đến 255.
Có 3 phương pháp chính để tính giá trị chuyển đổi:
*   Lightness: Tính giá trị trung bình của giá trị lớn nhất và nhỏ nhất cho các kênh màu: (max(R,G,B) + min(R,G,B))/2
*   Average: Tính giá trị trung bình của 3 kênh màu: (R+G+B)/3
*   Luminosity: Nhân hệ số tương ứng của 3 kênh màu như sau: 0.21*R + 0.72*G + 0.07*B

```python
import matplotlib.image as mpimg
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = mpimg.imread('/content/dog.jpeg')
plt.imshow(img)
```

### **Method 1: Lightness**

```python
def get_max(vector):
  result = vector.max()
  return result

def get_min(vector):
  result = vector.min()
  return result

gray_img_01 = (np.apply_along_axis(get_max, 2, img) + np.apply_along_axis(get_min, 2, img)) / 2
print(gray_img_01[0,0])
plt.imshow(gray_img_01, cmap='gray')
```

### **Method 2: Average**

```python
def get_average(vector):
  result = vector.mean()
  return result

gray_img_02 = np.apply_along_axis(get_average, 2, img)
print(gray_img_02[0,0])
plt.imshow(gray_img_02, cmap='gray')
```

### **Method 3: Luminosity**

```python
def get_luminosity(vector):
  result = 0.21 * vector[0] + 0.72 * vector[1] + 0.07 * vector[2]
  return result

gray_img_03 = np.apply_along_axis(get_luminosity, 2, img)
print(gray_img_03[0,0])
plt.imshow(gray_img_03, cmap='gray')
```
